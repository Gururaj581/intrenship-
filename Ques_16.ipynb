```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Create subplots: 1 row, 3 columns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Bar plot on the first subplot
axes[0].bar(categories, values, color='skyblue', edgecolor='black')
axes[0].set_title('Bar Plot')
axes[0].set_xlabel('Categories')
axes[0].set_ylabel('Values')

# Line plot on the second subplot
axes[1].plot(x, y, color='green', linewidth=2)
axes[1].set_title('Line Plot')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Sin(x)')

# Scatter plot on the third subplot
axes[2].scatter(x, z, color='red', s=30)
axes[2].set_title('Scatter Plot')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Cos(x)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
