```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example height vs weight data
data = {
    'Height': [150, 160, 165, 170, 180],
    'Weight': [50, 60, 65, 70, 80],
    'Size': [10, 15, 13, 18, 12]  # size for each point
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a scatter plot with variable point sizes
sns.scatterplot(
    x='Height',
    y='Weight',
    size='Size',
    data=df,
    sizes=(50, 300),  # control min and max point size
    legend=False,     # hide size legend if not needed
    color='blue'      # optional color
)

# Add a title and labels
plt.title('Height vs Weight (Point Size Based on Size Data)', fontsize=14)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')

# Show the plot
plt.show()
```
