```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset from Seaborn
iris = sns.load_dataset('iris')

# Create the boxplot for petal length by species
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set2')

# Add a title and labels
plt.title('Boxplot: Petal Length by Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')

# Show the plot
plt.show()

```
