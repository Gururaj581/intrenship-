```python
# Customizing Matplotlib Plot Styles
# Description: Customize the Matplotlib bar plot from Question 1 with a grid, different color, and custom fonts.

# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Example sales data
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [150, 200, 300, 100]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set a style (optional)
plt.style.use('seaborn-v0_8-darkgrid')

# Create the bar plot
plt.bar(df['Product'], df['Sales'], color='skyblue', edgecolor='black')

# Add a grid
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

# Customize fonts for title and labels
plt.title('Sales by Product', fontsize=16, fontweight='bold', fontname='Comic Sans MS')
plt.xlabel('Product', fontsize=12, fontname='Comic Sans MS')
plt.ylabel('Sales', fontsize=12, fontname='Comic Sans MS')

# Show the plot
plt.show()
```
