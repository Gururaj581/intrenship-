```python
# Seaborn Plot - Heatmap
# Objective: Create a heatmap to visualize a correlation matrix with Seaborn, using
# advanced styling.
# Import necessary libraries

# Import necessary libraries
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'tips' dataset from Seaborn
tips = sns.load_dataset('tips')

# Select only numeric columns for correlation calculation
numeric_columns = tips.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix of the numeric columns
corr_matrix = numeric_columns.corr()

# Create the heatmap with advanced styling
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(corr_matrix, 
            annot=True,           # Annotate cells with the numeric values
            cmap='coolwarm',      # Color palette for the heatmap
            fmt='.2f',            # Format numbers to two decimal places
            linewidths=0.5,       # Line width between cells
            linecolor='gray',     # Color of the lines between cells
            square=True,          # Make the heatmap square
            cbar_kws={'shrink': 0.75})  # Shrink the color bar slightly

# Add a title
plt.title('Correlation Matrix Heatmap', fontsize=16)

# Show the plot
plt.show()

```
