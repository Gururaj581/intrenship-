```python
# Basic Seaborn Scatter Plot
# Description: Using Seaborn, recreate the scatter plot from Question 2 for height vs. weight.

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Example height vs weight data
data = {
    'Height': [150, 160, 165, 170, 175, 180, 185],
    'Weight': [50, 60, 65, 70, 75, 80, 85]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a scatter plot
sns.scatterplot(x='Height', y='Weight', data=df)

# Add a title
plt.title('Height vs Weight')

# Show the plot
plt.show()
```
