```python
# Matplotlib Subplots for Multiple Data Sets
# Description: Create subplots using Matplotlib to display the bar plot for product sales and the scatter plot for
# height vs. weight side by side.
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Example data for product sales (Bar plot)
sales_data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [150, 200, 300, 100]
}
df_sales = pd.DataFrame(sales_data)

# Example data for height vs weight (Scatter plot)
height_weight_data = {
    'Height': [150, 160, 165, 170, 180],
    'Weight': [50, 60, 65, 70, 80]
}
df_height_weight = pd.DataFrame(height_weight_data)

# Create subplots: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar plot for sales data (subplot 1)
axes[0].bar(df_sales['Product'], df_sales['Sales'], color='skyblue', edgecolor='black')
axes[0].set_title('Product Sales')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sales')

# Scatter plot for height vs weight (subplot 2)
axes[1].scatter(df_height_weight['Height'], df_height_weight['Weight'], color='green')
axes[1].set_title('Height vs Weight')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```
