```python
# Advanced Seaborn Plot with DataFrame
# Description: Create a Seaborn plot using a DataFrame that visualizes two features in a single plot. Use tips
# dataset available in Seaborn's sample datasets.

# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips = sns.load_dataset('tips')

# Create a scatter plot with 'total_bill' on x-axis, 'tip' on y-axis, and hue based on 'day'
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', palette='Set1', size='size', sizes=(50, 200))

# Add a title and labels
plt.title('Scatter Plot: Total Bill vs Tip (with Day and Size)', fontsize=14)
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Show the plot
plt.show()
```
