```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Example data: x and y values for the line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot with custom styling
plt.plot(x, y, color='darkblue', linestyle='--', linewidth=2, marker='o', markersize=6, markerfacecolor='red', markeredgewidth=2)

# Add a title and labels
plt.title('Customized Line Plot: Sine Wave', fontsize=16, fontweight='bold')
plt.xlabel('X Axis (Time)', fontsize=12)
plt.ylabel('Y Axis (Amplitude)', fontsize=12)

# Customize grid
plt.grid(True, linestyle=':', color='gray', linewidth=0.5)

# Show the plot
plt.show()
```
