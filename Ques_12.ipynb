```python
# Advanced Seaborn Plot - Pairplot
# Objective: Use Seaborn to create a pairplot to visualize pairwise relationships in a dataset.
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips = sns.load_dataset('tips')

# Create a pairplot to visualize pairwise relationships in the dataset
sns.pairplot(tips, hue='sex', palette='coolwarm')

# Add a title
plt.suptitle('Pairplot: Pairwise Relationships in the Tips Dataset', fontsize=16)

# Show the plot
plt.show()
```
